```python
threshold = 200  # Adjust this threshold value if necessary
binary_image = np.where(np.sum(photo, axis=2) > threshold * 3, 1, 0)
plt.imshow(binary_image)
```

```python
percentage_white1 = (np.sum(binary_image) / binary_image.size) * 100
percentage_white1
```

```python
threshold2 = 200  # Adjust this threshold value if necessary
binary_image2 = np.where(np.sum(photo2, axis=2) > threshold2 * 3, 1, 0)
plt.imshow(binary_image2)
```

```python
percentage_white2 = (np.sum(binary_image2) / binary_image2.size) * 100
percentage_white2
```
